##Задача1:
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

```python
now = {'Apple': 1, 'Peach': 2, 'Plum': 3}  #товар на складе
privoz = {'Apple': 4, 'Peach': 5, 'Cherry': 6}    #товар привезли

def Sklad (dict1, dict2):    #функция, кол-во товаров на складе после привоза   
  for k, v in dict2.items():
    if k in dict1.keys():
      dict1[k] = v + dict1.get(k)
    else:
      dict1[k] = v
  return(dict1)

print(Sklad(now,privoz))
```

##Задача2:
Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200] <br>
Используйте list comprehensions

```python
cost = [100, 125, -90, 345, 655, -1, 0, 200]
result = [n for n in cost if n >= 0]
print(result)
print(f'Суммарные затраты = {sum(result)}')
```

###4.2 Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
def month_rev(list1, list2):
  rev2 = dict()                #создаем словарь
  for i in range(len(list2)):
    rev2[list1[i]] = list2[i]
  
  rev3 = dict()
  for k, v in rev2.items():
    if k[:7] in rev3.keys():    #если месяц уже есть в rev3, то суммируем values, иначе создаем новый key
      rev3[k[:7]] = rev3.get(k[:7])+rev2.get(k)
    else:
      rev3[k[:7]] = v
  return rev3

print(month_rev(data,summ))
```
